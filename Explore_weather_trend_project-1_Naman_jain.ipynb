```python
# Importing the Libraries 

import numpy as np
import pandas as pd                     # Loading data on Notebook
from matplotlib import pyplot as plt    # For Drawing chart
```

```python
# importing the  data which i download in form of csv file by performing some sql queries
data = pd.read_csv("results_main.csv")
```

```python
# Function that claculates moving Average  

def moving_avgerage(movingAverage_range, data_input):   # with local variables
    output = data_input.rolling(window = movingAverage_range, center = False, on = "cityat").mean().dropna()
    return output

#Calling the function with the range of Moving Average

movingAverage_value = 180

chartOf_moving_avg = moving_avgerage(movingAverage_value, data)     # with global variables
```

```python
# Drawing the graph of  Global Temperature

plt.plot(chartOf_moving_avg['year'], chartOf_moving_avg['globalat'], label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature in °C")
plt.title("Global Average Temperature")
plt.show()
```

```python
# Drawing the graph Temperature of Agra and Global temperature

plt.plot(chartOf_moving_avg['year'], chartOf_moving_avg['cityat'], label='Agra')
plt.plot(chartOf_moving_avg['year'], chartOf_moving_avg['globalat'], label='Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature in °C)")
plt.title("Agra Vs Global Average Temperature")
plt.show()
```
